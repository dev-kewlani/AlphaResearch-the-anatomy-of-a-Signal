Train-test split

```python
# Create forward-looking target
df['target_forward_return'] = df['log_return'].shift(-1)
df_clean = df.dropna(subset=['target_forward_return'])

# Get all feature columns
exclude_cols = ['mid_prc', 'price_change', 'price_return', 'log_return', 'target_forward_return', 'ret_1s_fwd']
feature_cols = [col for col in df_clean.columns if col not in exclude_cols]

print(f"\nTotal samples: {len(df_clean):,}")
print(f"Total features: {len(feature_cols)}")

# Time series split
train_pct = 0.7
train_samples = int(train_pct * len(df_clean))

X_train = df_clean[feature_cols].iloc[:train_samples]
y_train = df_clean['target_forward_return'].iloc[:train_samples]
X_test = df_clean[feature_cols].iloc[train_samples:]
y_test = df_clean['target_forward_return'].iloc[train_samples:]

print(f"\nTrain samples: {len(X_train):,} ({train_pct*100:.0f}%)")
print(f"Test samples: {len(X_test):,} ({(1-train_pct)*100:.0f}%)")
print(f"\nTrain target distribution:")
print(f"  Mean return: {y_train.mean():.6f}%")
print(f"  Std return: {y_train.std():.6f}%")
print(f"  Positive: {(y_train > 0).sum():,} ({(y_train > 0).mean()*100:.1f}%)")
print(f"  Negative: {(y_train < 0).sum():,} ({(y_train < 0).mean()*100:.1f}%)")
```

```python
from scipy.stats import norm, jarque_bera

df_train = df_clean.iloc[:train_samples]
df_test = df_clean.iloc[train_samples:]

print(f"Training samples: {len(df_train):,} ({train_pct*100:.0f}%)")
print(f"Test samples: {len(df_test):,} ({(1-train_pct)*100:.0f}%)")
print("="*60)

# Use log_return from training data only
r = df_train["log_return"].dropna().values
mu = r.mean()
sd = r.std(ddof=1)
z = (r - mu) / sd

print(f"\nTraining Data Statistics:")
print(f"Observations: {len(r):,}")
print(f"Mean: {mu:.8f}")
print(f"Std: {sd:.8f}")
print(f"Skew: {st.skew(r):.4f}")
print(f"Excess Kurtosis: {st.kurtosis(r):.4f}")

# Jarque-Bera normality test
jb_stat, jb_pvalue = jarque_bera(r)
print(f"\nJarque-Bera test:")
print(f"  Statistic: {jb_stat:.2f}")
print(f"  p-value: {jb_pvalue:.2e}")
print(f"  Result: {'REJECT normality' if jb_pvalue < 0.01 else 'Accept normality'}")

# Calculate tail ratios
print("\n" + "-"*60)
print("TAIL ANALYSIS")
print("-"*60)

tail_data = []
for t in (2, 3, 4):
    up = (z > t).mean() * 100
    dn = (z < -t).mean() * 100
    ng = (1 - norm.cdf(t)) * 100
    
    up_ratio = up / ng if ng > 0 else 0
    dn_ratio = dn / ng if ng > 0 else 0
    avg_ratio = (up_ratio + dn_ratio) / 2
    
    tail_data.append((t, up, dn, avg_ratio))
    
    print(f"{t}σ tails:")
    print(f"  Upper: {up:.3f}% (x{up_ratio:.1f} vs normal)")
    print(f"  Lower: {dn:.3f}% (x{dn_ratio:.1f} vs normal)")
    print(f"  Average ratio: {avg_ratio:.1f}")

# Overall assessment
avg_ratio = np.mean([ratio for _, _, _, ratio in tail_data])
print(f"\nOverall average tail ratio: {avg_ratio:.2f}")

if avg_ratio > 2:
    print("Assessment: SEVERE FAT TAILS")
elif avg_ratio > 1.5:
    print("Assessment: MODERATE FAT TAILS")
else:
    print("Assessment: Approximately normal tails")


# VISUALIZATIONS
fig, (ax1, ax2, ax4) = plt.subplots(1, 3, figsize=(18, 6))

# 1. LOG-SCALE HISTOGRAM
counts, bins, _ = ax1.hist(r, bins=100, density=True, alpha=0.7, 
                            color='lightblue', edgecolor='black', label='Training Data')
x = np.linspace(r.min(), r.max(), 1000)
normal_pdf = norm.pdf(x, mu, sd)
ax1.plot(x, normal_pdf, 'r-', linewidth=2, label='Normal (μ, σ)')
ax1.set_yscale('log')
ax1.set_ylim(bottom=1e-4)
ax1.set_title(f'Log-Scale Distribution (Training)\nExcess Kurtosis = {st.kurtosis(r):.2f}, JB p = {jb_pvalue:.2e}', 
              fontweight='bold', fontsize=11)
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('Log Returns')
ax1.set_ylabel('Density (Log Scale)')

# 2. Q-Q PLOT
st.probplot(r, dist="norm", plot=ax2)
ax2.set_title('Q-Q Plot vs Normal (Training)', fontweight='bold', fontsize=11)
ax2.grid(True, alpha=0.3)
ax2.get_lines()[0].set_markerfacecolor('blue')
ax2.get_lines()[0].set_alpha(0.7)
ax2.get_lines()[1].set_color('red')
ax2.get_lines()[1].set_linestyle('--')

# 3. TAIL PROBABILITY PLOT (CCDF)
tail_thresholds = np.linspace(1, 4.5, 100)
observed_tails = [(np.abs(z) > t).mean() for t in tail_thresholds]
normal_tails = [2 * (1 - norm.cdf(t)) for t in tail_thresholds]

ax4.semilogy(tail_thresholds, observed_tails, 'b-', linewidth=3, label='Observed (Training)')
ax4.semilogy(tail_thresholds, normal_tails, 'r--', linewidth=2, label='Normal')
ax4.fill_between(tail_thresholds, observed_tails, normal_tails, 
                 where=np.array(observed_tails) > np.array(normal_tails), 
                 alpha=0.3, color='red', label='Fat Tail Region')

# Mark key thresholds
for t in [2, 3, 4]:
    obs_prob = (np.abs(z) > t).mean()
    gauss_prob = 2 * (1 - norm.cdf(t))
    ratio = obs_prob / gauss_prob if gauss_prob > 0 else 0
    
    ax4.scatter([t], [obs_prob], s=50, color='darkblue', zorder=5)
    ax4.annotate(f'{obs_prob*100:.2f}% (x{ratio:.1f})', 
                xy=(t, obs_prob), xytext=(t+0.1, obs_prob*1.5),
                fontsize=9, ha='left', va='bottom',
                arrowprops=dict(arrowstyle='->', color='black', alpha=0.7))

ax4.set_xlabel('Standard Deviations |z|')
ax4.set_ylabel('Tail Probability P(|Z|>t)')
ax4.set_title('Tail Probability CCDF (Training)', fontweight='bold', fontsize=11)
ax4.legend()
ax4.grid(True, alpha=0.3)
ax4.set_xlim(1, 4.5)

plt.tight_layout()
plt.show()


for threshold in [1.5, 2.0, 2.5, 3.0, 3.5, 4.0]:
    obs_prob = (np.abs(z) > threshold).mean()
    norm_prob = 2 * (1 - norm.cdf(threshold))
    ratio = obs_prob / norm_prob if norm_prob > 0 else 0
    
    if ratio > 2:
        comment = "Very Fat"
    elif ratio > 1.5:
        comment = "Fat"
    elif ratio > 0.8:
        comment = "Normal"
    else:
        comment = "Thin"
    
    print(f"{threshold:>9.1f} | {obs_prob:>17.2e} | {norm_prob:>15.2e} | {ratio:>5.1f} | {comment}")

print()
print("METHODOLOGY")
print(f"• Analysis performed on TRAINING data only ({len(r):,} observations)")
print(f"• Normal overlays use sample mean (μ={mu:.6f}) and sample std (σ={sd:.6f})")
print(f"• Standardization: z = (r - μ)/σ with ddof=1 for unbiased estimator")
print(f"• Two-tailed probabilities: P(|Z| > t) for symmetric tail analysis")
print(f"• Fat tail threshold: Observed/Normal ratio > 1.5 indicates deviation")
print(f"• Test data ({len(df_test):,} obs) reserved for model validation only")

print()
print("TRAIN vs TEST DISTRIBUTION COMPARISON")

r_test = df_test["log_return"].dropna().values

print(f"\nTraining set:")
print(f"  Mean: {r.mean():.8f}, Std: {r.std():.8f}")
print(f"  Skew: {st.skew(r):.4f}, Kurtosis: {st.kurtosis(r):.4f}")

print(f"\nTest set:")
print(f"  Mean: {r_test.mean():.8f}, Std: {r_test.std():.8f}")
print(f"  Skew: {st.skew(r_test):.4f}, Kurtosis: {st.kurtosis(r_test):.4f}")
```

Volatility Regimes & Clustering Analysis

```python
# Calculate realized volatility on FULL dataset
window_300s = 300  # 5-minute window
df['realized_vol_5m'] = df['log_return'].rolling(window=window_300s, min_periods=60).std()

df_clean['realized_vol_5m'] = df['realized_vol_5m'].loc[df_clean.index]

df_train = df_clean.iloc[:train_samples]
df_test = df_clean.iloc[train_samples:]

print(f"\nCalculating volatility regimes using TRAINING data only...")
print(f"Training samples: {len(df_train):,}")
print(f"Test samples: {len(df_test):,}")

# Calculate regime thresholds using TRAINING data only
vol_measure_train = df_train['realized_vol_5m'].dropna()

regime_thresholds = {
    'low': vol_measure_train.quantile(0.33),
    'high': vol_measure_train.quantile(0.67)
}

print(f"\nVolatility Statistics (Training):")
print(f"  Mean: {vol_measure_train.mean():.6f}")
print(f"  Std: {vol_measure_train.std():.6f}")
print(f"  Low vol threshold (33rd percentile): ≤ {regime_thresholds['low']:.6f}")
print(f"  High vol threshold (67th percentile): ≥ {regime_thresholds['high']:.6f}")

# Apply regime classification to FULL dataset using TRAINING thresholds
df['vol_regime'] = df['realized_vol_5m'].apply(
    lambda x: 'Low' if pd.notna(x) and x <= regime_thresholds['low']
    else 'High' if pd.notna(x) and x >= regime_thresholds['high']
    else 'Medium' if pd.notna(x) else 'Unknown'
)

df_clean['vol_regime'] = df['vol_regime'].loc[df_clean.index]

df_train = df_clean.iloc[:train_samples]
df_test = df_clean.iloc[train_samples:]


abs_returns_train = df_train['log_return'].abs().dropna()
clustering_ac = abs_returns_train.autocorr(lag=1)

print(f"Lag-1 autocorr of |returns|: {clustering_ac:.4f}")
print(f"Clustering strength: {'Strong' if clustering_ac > 0.1 else 'Moderate' if clustering_ac > 0.05 else 'Weak'}")

# Ljung-Box test
try:
    from statsmodels.stats.diagnostic import acorr_ljungbox
    lb = acorr_ljungbox(abs_returns_train, lags=[10], return_df=True)
    lb_pvalue = lb['lb_pvalue'].iloc[0]
    print(f"\nLjung-Box test (lag 10):")
    print(f"  p-value: {lb_pvalue:.2e}")
    print(f"  Result: {'REJECT independence (clustering present)' if lb_pvalue < 0.01 else 'Cannot reject independence'}")
except ImportError:
    print(f"\nLjung-Box test: statsmodels not available")


regime_transitions = pd.crosstab(
    df_train['vol_regime'].shift(1), 
    df_train['vol_regime'], 
    normalize='index'
)

persistence_scores = {}
for regime in ['Low', 'Medium', 'High']:
    if regime in regime_transitions.index and regime in regime_transitions.columns:
        persistence_scores[regime] = regime_transitions.loc[regime, regime]
        print(f"{regime} regime persistence: {persistence_scores[regime]:.1%}")

if persistence_scores:
    avg_persistence = np.mean(list(persistence_scores.values()))
    print(f"\nAverage regime persistence: {avg_persistence:.1%}")

def spell_lengths(s, tag):
    idx = (s == tag).astype(int)
    cuts = (idx.diff() != 0).cumsum()
    spells = idx.groupby(cuts).sum()
    return spells.loc[idx.groupby(cuts).first() == 1]

print(f"\nRegime Duration Analysis (Training):")
regime_durations = {}
for tag in ['Low', 'Medium', 'High']:
    if tag in df_train['vol_regime'].values:
        lengths = spell_lengths(df_train['vol_regime'], tag)
        if len(lengths) > 0:
            regime_durations[tag] = lengths
            print(f"  {tag}: avg={lengths.mean():.1f}s, median={lengths.median():.0f}s, max={lengths.max()}s")


regime_counts_train = df_train['vol_regime'].value_counts()
regime_pcts_train = df_train['vol_regime'].value_counts(normalize=True).mul(100)

regime_counts_test = df_test['vol_regime'].value_counts()
regime_pcts_test = df_test['vol_regime'].value_counts(normalize=True).mul(100)

print(f"\nTraining Set:")
for regime in ['Low', 'Medium', 'High']:
    if regime in regime_counts_train:
        print(f"  {regime}: {regime_pcts_train[regime]:.1f}% ({regime_counts_train[regime]:,} obs)")

print(f"\nTest Set (for comparison):")
for regime in ['Low', 'Medium', 'High']:
    if regime in regime_counts_test:
        print(f"  {regime}: {regime_pcts_test[regime]:.1f}% ({regime_counts_test[regime]:,} obs)")

# Check if test distribution differs significantly
print("\nRegime Shift Check:")
for regime in ['Low', 'Medium', 'High']:
    if regime in regime_pcts_train and regime in regime_pcts_test:
        diff = regime_pcts_test[regime] - regime_pcts_train[regime]
        print(f"  {regime}: {diff:+.1f} percentage points")


fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Volatility over time with regime thresholds
ax1 = axes[0, 0]
vol_data = df['realized_vol_5m'].dropna()
train_vol = vol_data.iloc[:train_samples]
test_vol = vol_data.iloc[train_samples:]

ax1.plot(train_vol.index, train_vol, color='blue', linewidth=0.5, alpha=0.8, label='Training')
ax1.plot(test_vol.index, test_vol, color='orange', linewidth=0.5, alpha=0.8, label='Test')
ax1.axhline(y=regime_thresholds['low'], color='green', linestyle='--', alpha=0.8, label='Low threshold')
ax1.axhline(y=regime_thresholds['high'], color='red', linestyle='--', alpha=0.8, label='High threshold')
ax1.axvline(x=train_vol.index[-1], color='black', linestyle=':', linewidth=2, alpha=0.5)
ax1.set_ylabel('5-min Realized Volatility')
ax1.set_title('Volatility Regimes Over Time (Train/Test Split)', fontweight='bold')
ax1.legend(fontsize=8)
ax1.grid(True, alpha=0.3)

# Plot 2: Autocorrelation of absolute returns
ax2 = axes[0, 1]
lags = range(1, 21)
autocorrs = [abs_returns_train.autocorr(lag=lag) for lag in lags]
ax2.plot(lags, autocorrs, 'b-', linewidth=2, marker='o', markersize=4)
ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3)
ax2.axhline(y=2/np.sqrt(len(abs_returns_train)), color='red', linestyle='--', 
            alpha=0.7, label='95% confidence')
ax2.axhline(y=-2/np.sqrt(len(abs_returns_train)), color='red', linestyle='--', alpha=0.7)
ax2.set_xlabel('Lag')
ax2.set_ylabel('Autocorrelation')
ax2.set_title('Volatility Clustering (Training)', fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Regime distribution comparison
ax3 = axes[1, 0]
regimes = ['Low', 'Medium', 'High']
train_pcts = [regime_pcts_train.get(r, 0) for r in regimes]
test_pcts = [regime_pcts_test.get(r, 0) for r in regimes]

x = np.arange(len(regimes))
width = 0.35
ax3.bar(x - width/2, train_pcts, width, label='Training', color='steelblue', edgecolor='black')
ax3.bar(x + width/2, test_pcts, width, label='Test', color='orange', edgecolor='black')
ax3.set_ylabel('Percentage (%)')
ax3.set_title('Regime Distribution: Train vs Test', fontweight='bold')
ax3.set_xticks(x)
ax3.set_xticklabels(regimes)
ax3.legend()
ax3.grid(True, alpha=0.3, axis='y')

# Plot 4: Regime transition matrix
ax4 = axes[1, 1]
if len(regime_transitions) > 0:
    im = ax4.imshow(regime_transitions.values, cmap='Blues', aspect='auto', vmin=0, vmax=1)
    ax4.set_xticks(range(len(regime_transitions.columns)))
    ax4.set_yticks(range(len(regime_transitions.index)))
    ax4.set_xticklabels(regime_transitions.columns)
    ax4.set_yticklabels(regime_transitions.index)
    ax4.set_xlabel('To Regime')
    ax4.set_ylabel('From Regime')
    ax4.set_title('Regime Transition Probabilities (Training)', fontweight='bold')
   
    for i in range(len(regime_transitions.index)):
        for j in range(len(regime_transitions.columns)):
            text = ax4.text(j, i, f'{regime_transitions.iloc[i, j]:.2f}',
                          ha="center", va="center", color="black", fontsize=10)
    
    plt.colorbar(im, ax=ax4, label='Probability')

plt.tight_layout()
plt.show()
```

Feature Correlation Analysis

```python

df_train = df_clean.iloc[:train_samples].copy()

all_cols = df_train.columns.tolist()
engineered_markers = ['logp', 'ret_1s', 'price_change', 'log_return', 'hour', 'target_forward_return']
engineered_start_idx = min([all_cols.index(m) for m in engineered_markers if m in all_cols], default=len(all_cols))

original_cols = all_cols[:engineered_start_idx]
feature_cols = df_train[original_cols].select_dtypes(include=[np.number]).columns.tolist()
feature_cols = [col for col in feature_cols if col != 'mid_prc' and df_train[col].std() > 1e-10]


# 1. FEATURE-TO-TARGET CORRELATION (SPEARMAN)
from scipy.stats import spearmanr

target_corrs = []
for feature in feature_cols:
    valid_mask = ~(df_train[feature].isna() | df_train['log_return'].isna())
    if valid_mask.sum() > 1:
        corr, _ = spearmanr(df_train[feature][valid_mask], df_train['log_return'][valid_mask])
        target_corrs.append((feature, corr))

target_corrs_df = pd.DataFrame(target_corrs, columns=['feature', 'correlation'])
target_corrs_df['abs_corr'] = target_corrs_df['correlation'].abs()
target_corrs_df = target_corrs_df.sort_values('abs_corr', ascending=False)

print(f"\nTop 15 features by Spearman correlation with log_return:")
for i, row in enumerate(target_corrs_df.head(15).itertuples(), 1):
    print(f"{i:2d}. {row.feature:<30} {row.correlation:>8.4f}")

# Key statistics
sig_10 = (target_corrs_df['abs_corr'] > 0.10).sum()
sig_05 = (target_corrs_df['abs_corr'] > 0.05).sum()
print(f"\nFeatures with |ρ| > 0.10: {sig_10} ({sig_10/len(feature_cols)*100:.1f}%)")
print(f"Features with |ρ| > 0.05: {sig_05} ({sig_05/len(feature_cols)*100:.1f}%)")

# Visualize
top_n = min(25, len(target_corrs_df))
plt.figure(figsize=(10, 8))
top_features = target_corrs_df.head(top_n)
colors = ['green' if x > 0 else 'red' for x in top_features['correlation']]
plt.barh(range(top_n), top_features['correlation'], color=colors, alpha=0.7)
plt.yticks(range(top_n), top_features['feature'], fontsize=9)
plt.xlabel('Spearman Correlation with log_return')
plt.title('Top Features by Target Correlation (Spearman, Training)', fontweight='bold')
plt.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
plt.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.show()

# 2. FEATURE-TO-FEATURE REDUNDANCY (SPEARMAN)
print("\n2. FEATURE REDUNDANCY (SPEARMAN)")
print("-" * 60)

# Calculate Spearman correlation matrix
corr_matrix = df_train[feature_cols].corr(method='spearman')
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
all_corrs = upper_triangle.stack()

# Find highly correlated pairs
high_corr_pairs = all_corrs[all_corrs.abs() > 0.95].sort_values(key=abs, ascending=False)

print(f"Highly correlated pairs (|ρ| > 0.95): {len(high_corr_pairs)}")
if len(high_corr_pairs) > 0:
    print("\nTop 10 redundant pairs:")
    for i, (pair, corr) in enumerate(high_corr_pairs.head(10).items(), 1):
        print(f"{i:2d}. {pair[0]} - {pair[1]}: {corr:+.4f}")

# Summary statistics
abs_corrs = all_corrs.abs().values
print(f"\nCorrelation statistics:")
print(f"  Mean |correlation|: {abs_corrs.mean():.4f}")
print(f"  Max |correlation|: {abs_corrs.max():.4f}")
print(f"  Pairs with |ρ| > 0.80: {(abs_corrs > 0.80).sum()}")
print(f"  Pairs with |ρ| > 0.95: {(abs_corrs > 0.95).sum()}")

# 3. FEATURE VARIANCE
print("\n3. FEATURE VARIANCE")
print("-" * 60)

feature_vars = df_train[feature_cols].var().sort_values(ascending=False)

print("Top 10 highest variance features:")
for i, (feat, var_val) in enumerate(feature_vars.head(10).items(), 1):
    print(f"{i:2d}. {feat}: {var_val:.4f}")

print(f"\nVariance statistics:")
print(f"  Mean: {feature_vars.mean():.4f}")
print(f"  Max/Min ratio: {feature_vars.max() / feature_vars.min():.1f}x")

# 4. REDUNDANCY REMOVAL

features_to_drop = set()
target_corr_dict = dict(zip(target_corrs_df['feature'], target_corrs_df['abs_corr']))

for (feat1, feat2), corr_val in high_corr_pairs.items():
    if feat1 not in features_to_drop and feat2 not in features_to_drop:
        corr1 = target_corr_dict.get(feat1, 0)
        corr2 = target_corr_dict.get(feat2, 0)
        
        if corr1 >= corr2:
            features_to_drop.add(feat2)
        else:
            features_to_drop.add(feat1)

print(f"Original features: {len(feature_cols)}")
print(f"Features to drop: {len(features_to_drop)}")
print(f"Final features: {len(feature_cols) - len(features_to_drop)}")

# Create clean feature list
feature_cols_clean = [f for f in feature_cols if f not in features_to_drop]

# Verify reduction in redundancy
corr_matrix_clean = df_train[feature_cols_clean].corr(method='spearman')
upper_clean = corr_matrix_clean.where(np.triu(np.ones(corr_matrix_clean.shape), k=1).astype(bool))
high_corr_clean = (upper_clean.stack().abs() > 0.95).sum()

print(f"\nRedundancy check:")
print(f"  Before: {len(high_corr_pairs)} pairs with |ρ| > 0.95")
print(f"  After: {high_corr_clean} pairs with |ρ| > 0.95")
print(f"  Reduction: {(1 - high_corr_clean/len(high_corr_pairs))*100:.1f}%")
```

Signal Decay Analysis

```python
df_train = df_clean.iloc[:train_samples].copy()

# Calculate forward returns at different horizons
max_horizon = 20
for h in range(1, max_horizon + 1):
    df_train[f'fwd_return_{h}'] = df_train['log_return'].shift(-h)

print(f"Analyzing {len(feature_cols_clean)} cleaned features\n")

# Calculate Spearman correlation at each horizon
from scipy.stats import spearmanr

decay_results = []
for feature in feature_cols_clean:
    correlations = []
    for h in range(1, max_horizon + 1):
        valid_mask = ~(df_train[feature].isna() | df_train[f'fwd_return_{h}'].isna())
        if valid_mask.sum() > 10:
            corr, _ = spearmanr(df_train[feature][valid_mask], 
                               df_train[f'fwd_return_{h}'][valid_mask])
            correlations.append(corr)
        else:
            correlations.append(np.nan)
    
    # Calculate half-life
    initial = abs(correlations[0])
    half_point = initial * 0.5
    half_life = max_horizon
    for h, corr in enumerate(correlations):
        if abs(corr) <= half_point:
            half_life = h + 1
            break
    
    decay_results.append({
        'feature': feature,
        'correlations': correlations,
        'initial_corr': correlations[0],
        'half_life': half_life
    })

decay_df = pd.DataFrame(decay_results)
median_half_life = decay_df['half_life'].median()

print(f"Median half-life: {median_half_life:.0f} periods")
print(f"Half-life range: {decay_df['half_life'].min():.0f} - {decay_df['half_life'].max():.0f} periods")

# Visualizations
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Average decay
horizons = list(range(1, max_horizon + 1))
avg_abs_corr = np.nanmean([np.abs(r['correlations']) for r in decay_results], axis=0)

ax1.plot(horizons, avg_abs_corr, 'o-', linewidth=2, markersize=6, color='steelblue')
ax1.axhline(y=avg_abs_corr[0] * 0.5, color='red', linestyle='--', alpha=0.5, label='50% decay')
ax1.axvline(x=median_half_life, color='orange', linestyle='--', alpha=0.5, label=f'Median half-life ({median_half_life:.0f})')
ax1.set_xlabel('Horizon (periods ahead)')
ax1.set_ylabel('Average |Spearman ρ|')
ax1.set_title('Average Signal Decay', fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Top 5 features
decay_df_sorted = decay_df.sort_values('initial_corr', key=abs, ascending=False)
for _, row in decay_df_sorted.head(5).iterrows():
    abs_corrs = [abs(c) for c in row['correlations']]
    ax2.plot(horizons, abs_corrs, 'o-', linewidth=1.5, markersize=4, 
             label=f"{row['feature'][:10]} (h={row['half_life']:.0f})", alpha=0.7)

ax2.set_xlabel('Horizon (periods ahead)')
ax2.set_ylabel('|Spearman ρ|')
ax2.set_title('Top 5 Features by Initial Correlation', fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.legend(fontsize=9)

plt.tight_layout()
plt.show()

# Key statistics
decay_rate = (avg_abs_corr[0] - avg_abs_corr[-1]) / avg_abs_corr[0] * 100

print(f"\nSignal characteristics:")
print(f"  Initial correlation: {avg_abs_corr[0]:.4f}")
print(f"  Final correlation: {avg_abs_corr[-1]:.4f}")
print(f"  Total decay: {decay_rate:.1f}%")
print(f"  Fast decay (≤5 periods): {(decay_df['half_life'] <= 5).sum()} features")
```

Extremelly fast signal decay

Half-life: 2 seconds
Effectively: 6-8 seconds
Full decay: 40 seconds

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

# Create binary classification target
y_train_class = (y_train > 0).astype(int)
y_test_class = (y_test > 0).astype(int)

print(f"\nClass distribution:")
print(f"  Train - Down: {(y_train_class == 0).mean()*100:.1f}%, Up: {(y_train_class == 1).mean()*100:.1f}%")
print(f"  Test - Down: {(y_test_class == 0).mean()*100:.1f}%, Up: {(y_test_class == 1).mean()*100:.1f}%")

# Train Random Forest Classifier
rf_classifier = RandomForestClassifier(
    n_estimators=200,
    max_depth=8,
    min_samples_split=100,
    min_samples_leaf=50,
    max_features='sqrt',
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

print("\nTraining Random Forest Classifier...")
rf_classifier.fit(X_train[feature_cols_clean], y_train_class)
print("Training complete!")

# Generate predictions
y_test_proba = rf_classifier.predict_proba(X_test[feature_cols_clean])[:, 1]
test_auc = roc_auc_score(y_test_class, y_test_proba)

print(f"\nModel Performance:")
print(f"  ROC-AUC: {test_auc:.4f}")
print(f"  Probability range: [{y_test_proba.min():.3f}, {y_test_proba.max():.3f}]")

# STEP 2: IMPLEMENT STRATEGY WITH ENTRY/EXIT LOGIC
print("\n" + "="*60)
print("FINAL TRADING STRATEGY")
print("Long & Short with Entry/Exit Thresholds")
print("="*60)

# Create test results dataframe
test_results = df_clean.iloc[train_samples:].copy()
test_results['predicted_prob'] = y_test_proba

# Cap extreme returns
return_cap = 0.05
test_results['target_return_capped'] = test_results['target_forward_return'].clip(-return_cap, return_cap)

# Generate signals with long/short/flat logic
long_entry = 0.68   # Enter long when prob > 0.68
long_exit = 0.60    # Exit long when prob < 0.60
short_entry = 0.32  # Enter short when prob < 0.32
short_exit = 0.40   # Exit short when prob > 0.40

position = 0
signals = []

for i in range(len(test_results)):
    prob = test_results['predicted_prob'].iloc[i]
    
    # If flat, check for entry signals
    if position == 0:
        if prob > long_entry:
            position = 1  # Go long
        elif prob < short_entry:
            position = -1  # Go short
    
    # If long, check for exit
    elif position == 1:
        if prob < long_exit:
            position = 0  # Exit to flat
    
    # If short, check for exit
    elif position == -1:
        if prob > short_exit:
            position = 0  # Exit to flat
    
    signals.append(position)

test_results['signal_final'] = signals

# Verify position changes
n_changes = (test_results['signal_final'].diff() != 0).sum()
n_long = (test_results['signal_final'] == 1).sum()
n_short = (test_results['signal_final'] == -1).sum()
n_flat = (test_results['signal_final'] == 0).sum()

print(f"\nSignal Generation:")
print(f"  Long entry: prob > {long_entry}, exit: prob < {long_exit}")
print(f"  Short entry: prob < {short_entry}, exit: prob > {short_exit}")
print(f"  Position changes: {n_changes:,}")
print(f"  Long periods: {n_long:,} ({n_long/len(test_results)*100:.1f}%)")
print(f"  Short periods: {n_short:,} ({n_short/len(test_results)*100:.1f}%)")
print(f"  Flat periods: {n_flat:,} ({n_flat/len(test_results)*100:.1f}%)")

# Calculate returns
test_results['strategy_return_final'] = test_results['signal_final'] * test_results['target_return_capped']
test_results['log_strat_final'] = test_results['strategy_return_final'] / 100
test_results['log_market_final'] = test_results['target_return_capped'] / 100

# Cumulative returns
test_results['cumulative_strategy'] = (np.exp(test_results['log_strat_final'].cumsum()) - 1) * 100
test_results['cumulative_market'] = (np.exp(test_results['log_market_final'].cumsum()) - 1) * 100

# Get active trades
active_trades = test_results[test_results['signal_final'] != 0].copy()

# PERFORMANCE METRICS
print("\n--- Performance Summary ---")
print(f"Total Return: {test_results['cumulative_strategy'].iloc[-1]:.2f}%")
print(f"Market Return: {test_results['cumulative_market'].iloc[-1]:.2f}%")
print(f"Alpha (Outperformance): {test_results['cumulative_strategy'].iloc[-1] - test_results['cumulative_market'].iloc[-1]:.2f}%")

# Trading statistics
n_trades = len(active_trades)
pct_time_active = n_trades / len(test_results)
win_rate = (active_trades['strategy_return_final'] > 0).mean()
avg_win = active_trades[active_trades['strategy_return_final'] > 0]['strategy_return_final'].mean()
avg_loss = active_trades[active_trades['strategy_return_final'] < 0]['strategy_return_final'].mean()
sharpe = active_trades['strategy_return_final'].mean() / active_trades['strategy_return_final'].std() * np.sqrt(252)

print(f"\n--- Trading Statistics ---")
print(f"Number of Trades: {n_trades:,}")
print(f"Time in Market: {pct_time_active*100:.2f}%")
print(f"Win Rate: {win_rate*100:.2f}%")
print(f"Average Trade Return: {active_trades['strategy_return_final'].mean():.4f}%")
print(f"Average Win: {avg_win:.4f}%")
print(f"Average Loss: {avg_loss:.4f}%")
print(f"Profit Factor: {abs(avg_win/avg_loss):.2f}")
print(f"Sharpe Ratio (annualized): {sharpe:.2f}")

# DRAWDOWN ANALYSIS
print("\n--- Risk Metrics ---")

# Calculate running maximum and drawdown
running_max = test_results['cumulative_strategy'].expanding().max()
drawdown = (test_results['cumulative_strategy'] - running_max)
max_drawdown = drawdown.min()
max_dd_pct = (max_drawdown / (running_max[drawdown.idxmin()] + 100)) * 100 if running_max[drawdown.idxmin()] > 0 else 0

print(f"Maximum Drawdown: {max_drawdown:.2f}% ({max_dd_pct:.2f}% from peak)")
print(f"Standard Deviation (annual): {active_trades['strategy_return_final'].std() * np.sqrt(252):.4f}%")

# RETURN DISTRIBUTION
print("\n--- Return Distribution ---")
print(f"Best Trade: {active_trades['strategy_return_final'].max():.4f}%")
print(f"Worst Trade: {active_trades['strategy_return_final'].min():.4f}%")
print(f"Median Trade: {active_trades['strategy_return_final'].median():.4f}%")
print(f"25th Percentile: {active_trades['strategy_return_final'].quantile(0.25):.4f}%")
print(f"75th Percentile: {active_trades['strategy_return_final'].quantile(0.75):.4f}%")

# TRANSACTION COSTS ANALYSIS
print("\n" + "="*60)
print("TRANSACTION COSTS ANALYSIS")
print("="*60)

# Calculate position changes
test_results['position_change'] = test_results['signal_final'].diff().fillna(test_results['signal_final']).abs()

# Test different cost scenarios
cost_scenarios = [0.0001, 0.0003, 0.0005, 0.001, 0.002]  # 1, 3, 5, 10, 20 bps

print("\nNet Returns After Transaction Costs:")
print(f"{'Cost (bps)':<12} {'Net Return':<15} {'Net Sharpe':<12} {'Total Costs'}")
print("-" * 60)

for cost in cost_scenarios:
    test_results['costs'] = test_results['position_change'] * cost
    test_results['return_net'] = test_results['strategy_return_final'] - test_results['costs']
    test_results['log_net'] = test_results['return_net'] / 100
    cum_net = (np.exp(test_results['log_net'].cumsum()) - 1) * 100
    
    active_net = test_results[test_results['signal_final'] != 0].copy()
    sharpe_net = active_net['return_net'].mean() / active_net['return_net'].std() * np.sqrt(252)
    total_costs = test_results['costs'].sum()
    
    print(f"{cost*10000:<12.1f} {cum_net.iloc[-1]:<15.2f} {sharpe_net:<12.2f} {total_costs:.4f}%")

# VISUALIZATIONS
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# 1. Cumulative Returns
ax1 = axes[0, 0]
ax1.plot(test_results.index, test_results['cumulative_strategy'], label='Strategy', linewidth=2, color='green')
ax1.plot(test_results.index, test_results['cumulative_market'], label='Market', linewidth=2, color='gray', alpha=0.7)
ax1.set_title('Cumulative Returns: Strategy vs Market', fontsize=14, fontweight='bold')
ax1.set_xlabel('Time')
ax1.set_ylabel('Cumulative Return (%)')
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. Drawdown
ax2 = axes[0, 1]
ax2.fill_between(test_results.index, drawdown, 0, color='red', alpha=0.3)
ax2.plot(test_results.index, drawdown, color='darkred', linewidth=1)
ax2.set_title('Strategy Drawdown', fontsize=14, fontweight='bold')
ax2.set_xlabel('Time')
ax2.set_ylabel('Drawdown (%)')
ax2.grid(True, alpha=0.3)

# 3. Return Distribution
ax3 = axes[1, 0]
ax3.hist(active_trades['strategy_return_final'], bins=50, color='steelblue', alpha=0.7, edgecolor='black')
ax3.axvline(0, color='red', linestyle='--', linewidth=2, label='Break-even')
ax3.axvline(active_trades['strategy_return_final'].mean(), color='green', linestyle='--', linewidth=2, label='Mean')
ax3.set_title('Distribution of Trade Returns', fontsize=14, fontweight='bold')
ax3.set_xlabel('Return (%)')
ax3.set_ylabel('Frequency')
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. Net Position Over Time
ax4 = axes[1, 1]

# Plot each position state separately for clarity
long_mask = test_results['signal_final'] == 1
short_mask = test_results['signal_final'] == -1

ax4.fill_between(test_results.index, 1, 0, where=long_mask, 
                 color='green', alpha=0.5, label='Long', step='post')
ax4.fill_between(test_results.index, 0, -1, where=short_mask, 
                 color='red', alpha=0.5, label='Short', step='post')

ax4.set_title('Net Position Over Time', fontsize=14, fontweight='bold')
ax4.set_xlabel('Time')
ax4.set_ylabel('Position')
ax4.set_ylim([-1.2, 1.2])
ax4.set_yticks([-1, 0, 1])
ax4.set_yticklabels(['Short', 'Flat', 'Long'])
ax4.axhline(y=0, color='black', linewidth=0.5)
ax4.legend(loc='upper left')
ax4.grid(True, alpha=0.3)

# Create results summary
results_summary = pd.DataFrame({
    'Metric': [
        'Total Return (%)',
        'Market Return (%)',
        'Alpha (%)',
        'Sharpe Ratio',
        'Win Rate (%)',
        'Number of Trades',
        'Time in Market (%)',
        'Position Changes',
        'Max Drawdown (%)',
        'Avg Trade Return (%)',
        'Profit Factor'
    ],
    'Value': [
        f"{test_results['cumulative_strategy'].iloc[-1]:.2f}",
        f"{test_results['cumulative_market'].iloc[-1]:.2f}",
        f"{test_results['cumulative_strategy'].iloc[-1] - test_results['cumulative_market'].iloc[-1]:.2f}",
        f"{sharpe:.2f}",
        f"{win_rate*100:.2f}",
        f"{n_trades:,}",
        f"{pct_time_active*100:.2f}",
        f"{n_changes:,}",
        f"{max_drawdown:.2f}",
        f"{active_trades['strategy_return_final'].mean():.4f}",
        f"{abs(avg_win/avg_loss):.2f}"
    ]
})

print("\nStrategy Summary:")
print(results_summary.to_string(index=False))

# Save predictions and signals to the dataframe
test_results['model_probability'] = test_results['predicted_prob']
test_results['model_signal'] = test_results['signal_final']

print("\nColumns added to test_results:")
print("  - model_probability: Predicted probability of up move")
print("  - model_signal: Trading signal (1=Long, 0=Flat)")
```
